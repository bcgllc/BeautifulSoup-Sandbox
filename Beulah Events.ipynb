```python
from bs4 import BeautifulSoup
import requests
import json
import re
```

```python
request = requests.get('http://beulah.com/events/')
data = request.text
```

```python
soup = BeautifulSoup(data, 'html.parser')
```

```python
container = soup.find('div', {'class': 'tribe-events-loop'})
print(container)
```

```python
events = []
```

```python
eventItems = container.find_all('div', {'class': 'bord'})
```

```python
for index, eventItem in enumerate(eventItems):
    title = eventItem.find('a', {'class': 'tribe-event-url'})
    start_date = eventItem.find('span', {'class': 'tribe-event-date-start'})
    event = {}
    event['title'] = title.text
    events.append(event)
```

```python
print(events)
```

```python
json_data = json.dumps(events)
print(json_data)
```
