# Check Latest Prices for Ultrawide Monitors on newegg.com
## with Python, Jupyter, and BeautifulSoup

```python
from bs4 import BeautifulSoup
import requests
import json
```

```python
source_url = 'https://www.newegg.com/p/pl?N=100161617%20601276169&Order=BESTMATCH'
```

```python
request = requests.get(source_url)
data = request.text
```

```python
soup = BeautifulSoup(data, 'html.parser')
```

```python
items_view = soup.find('div', {'class': 'items-view is-grid'})
```

```python
items = items_view.find_all('div', {'class': 'item-container'})
print(len(items))
```

```python
featured_items = []
```

```python
for _, item in enumerate(items):
    item_image_url = item.find('a', {'class': 'item-img'}).find('img').attrs['src']
    item_url = item.find('a', {'class': 'item-img'}).attrs['href']
    item_title = item.find('a', {'class': 'item-title'}).text
    item_price_dollars = item.find('li', {'class': 'price-current'}).find('strong').text
    item_price_cents = item.find('li', {'class': 'price-current'}).find('sup').text
    item_savings_dollar = item.find('span', {'class': 'price-save-dollar'})
    item_savings_percent = item.find('span', {'class': 'price-save-percent'})
    
    featured_item = {}
    featured_item['item_image_url'] = item_image_url
    featured_item['item_url'] = item_url
    featured_item['item_title'] = item_title
    featured_item['item_price'] = '$' + item_price_dollars + item_price_cents
    featured_item['item_savings_dollar'] = '' if item_savings_dollar is None else item_savings_dollar.text
    featured_item['item_savings_percent'] = '' if item_savings_percent is None else item_savings_percent.text
    
    featured_items.append(featured_item)
```

```python
#print(len(featured_items))
#print(featured_items)

json_data = json.dumps(featured_items, indent=2, sort_keys=True)
print(json_data)
```
